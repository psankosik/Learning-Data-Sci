# EWMA

## MA 
### Moving Average

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('../../Data/airline_passengers.csv',index_col='Month')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df.head()
```

```python
df['6-month-SMA'] = df['Thousands of Passengers'].rolling(window=6).mean()
```

```python
df['12-month-SMA'] = df['Thousands of Passengers'].rolling(window=12).mean()
```

```python
df.plot(figsize=(20,5));
```

## EWMA 
### Exponential weighted moving avgerage

```python
df['EMWA-12'] = df['Thousands of Passengers'].ewm(span=12).mean()
```

```python
df[['Thousands of Passengers','EMWA-12']].plot(figsize=(20,5));
```
