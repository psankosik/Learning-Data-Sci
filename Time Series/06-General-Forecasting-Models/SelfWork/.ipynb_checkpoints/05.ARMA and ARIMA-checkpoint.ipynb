# ARMA($p,q$) and ARIMA($p,d,q$)

```python
from statsmodels.tsa.arima_model import ARMA,ARIMA,ARMAResults,ARIMAResults
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
from pmdarima import auto_arima
```

```python
df1 = pd.read_csv('../../Data/DailyTotalFemaleBirths.csv',index_col='Date',parse_dates=True)
df1.index.freq = 'D'
df1 = df1[:120]
```

```python
df2 = pd.read_csv('../../Data/TradeInventories.csv',index_col='Date',parse_dates=True)
df2.index.freq = 'MS'
```

## ARMA

```python
df1['Births'].plot(figsize=(20,8));
```

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
adf_test(df1['Births'])
```

```python
auto_arima(df1['Births'],start_p=0,start_q=0,seasonal=False,trace=True).summary()
```

```python
train = df1.iloc[:90]
test = df1.iloc[90:]
```

```python
model = ARMA(train['Births'].astype('float32'),order=(2,2))
```

```python
results = model.fit()
```

```python
results.summary()
```

```python
start = len(train)
end = len(train)+len(test)-1
```

```python
predictions = results.predict(start,end).rename('ARMA (2,2) Predictions')
```

```python
test['Births'].plot(figsize=(20,5),legend=True)
predictions.plot(legend = True)
```

```python
abs(test.mean()-predictions.mean())
```

```python
train = df2.iloc[:252]
test = df2.iloc[252:]
```

```python
results = ARIMA(train['Inventories'],order=(1,1,1)).fit()
results.summary()
```

```python
start = len(train)
end = len(train) + len(test) - 1

predictions = results.predict(start=start,end=end,typ='levels').rename('ARIMA(1,1,1) Predictions')
```

```python
test['Inventories'].plot(figsize=(20,8),legend=True)
predictions.plot(legend=True)
```

```python
from statsmodels.tools.eval_measures import rmse
```

```python
error = rmse(test['Inventories'],predictions)
```

## Forecast into Future

```python
results = ARIMA(df2['Inventories'],order=(1,1,1)).fit()
```

```python
fcast = results.predict(start=len(df2),end=len(df2)+11,typ='levels').rename('ARIMA (1,1,1) FORECAST')
```

```python
df2['Inventories'].plot(figsize=(20,8),legend=True)
fcast.plot(legend=True);
```
